```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d=pd.read_csv("Space_Missions_Cleaned.csv")
o=d.iloc[:,6]
b=d.iloc[:,0]
o=o.tolist()
```

```python
des=[]
for i in range(len(o)):
    if o[i]=='Success':
        des.append(1)
    else:
        des.append(0)
a1 = pd.DataFrame(b)        
des=pd.Series(des,name="success")
b1=pd.DataFrame(des)
des=pd.concat([a1,des], axis=1)
```

```python
des=des.groupby("Company Name")
f1=des.sum()
f=des.count()
g=f.iloc[:,0]
g=g.tolist()
g1=f1.iloc[:,0]
g1=g1.tolist()
ht=[]
for i in range(len(g)):
    ht.append(g1[i]/g[i]*100)
```

```python
b=b.tolist()
g={'0'}
for i in b:
    g.add(i)
g.discard('0')
```

```python
l=[]
for i in g:
    l.append(i)
l.sort()
ht2=pd.Series(ht,name="SP")
ht2=pd.DataFrame(ht2)
l1=pd.Series(l,name="Comany Name")
l1=pd.DataFrame(l1)
ht1=pd.concat([l1,ht2],axis=1)
```

```python
ht1=ht1.sort_values(["SP"], ascending=[0])
t=ht1.iloc[:10,:]
t
```

```python
g=t.iloc[:,0]
g=g.tolist()
f=t.iloc[:,1]
f=f.tolist()
```
